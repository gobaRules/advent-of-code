```python
with open("3.txt") as f:
    lines = f.read().splitlines()
```

```python
lines
```

```python
total = 0

for line in lines:
    half_length = int(len(line)/2)
    first_comp = line[0:half_length]
    second_comp = line[half_length:]
    print(first_comp, second_comp)
    
    for char in first_comp:
        if char in second_comp:
            print(char)
            break
    
    priority = 0   
    alphabet = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    priority += alphabet.find(char) + 1
    print(priority)
    
    total += priority
    
total
```

```python
with open("3.txt") as f:
    lines = f.read().splitlines()
lines
```

```python
total = 0

for i, line in enumerate(lines):
   
    if i%3 == 0:
        print(i)
        for char in line:
            if char in lines[i+1] and char in lines[i+2]:
                print(char)
                break
    
        priority = 0   
        alphabet = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
        priority += alphabet.find(char) + 1
        print(priority)
    
        total += priority

total
```
