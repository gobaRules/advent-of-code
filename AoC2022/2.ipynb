```python
with open("2.txt") as f:
    lines = f.read().splitlines()
```

```python
lines
```

```python
total = 0

for line in lines:
    op = line[0]
    me = line[-1]
    
    
    # I win = 6
    if op == "A" and me == "Y":
        total += 8
    if op == "B" and me == "Z":
        total += 9
    if op == "C" and me == "X":
        total += 7
        
    # I loose = 0
    if op == "A" and me == "Z":
        total += 3
    if op == "B" and me == "X":
        total += 1
    if op == "C" and me == "Y":
        total += 2
        
    # Draw = 3
    if op == "A" and me == "X":
        total += 4
    if op == "B" and me == "Y":
        total += 5
    if op == "C" and me == "Z":
        total += 6


total
```

```python
total = 0
# X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win

for line in lines:
    op = line[0]
    me = line[-1]
    
     
    # I win = 6
    if op == "A" and me == "Z":
        total += 8
    if op == "B" and me == "Z":
        total += 9
    if op == "C" and me == "Z":
        total += 7
        
    # I loose = 0
    if op == "A" and me == "X":
        total += 3
    if op == "B" and me == "X":
        total += 1
    if op == "C" and me == "X":
        total += 2
        
    # Draw = 3
    if op == "A" and me == "Y":
        total += 4
    if op == "B" and me == "Y":
        total += 5
    if op == "C" and me == "Y":
        total += 6
        
total
```
